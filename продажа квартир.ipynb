Цену продажи переведем в тысячи рублей:

### Определение неуказанных населенных пунктов по расстоянию до центра и до аэропорта: 

````python
location_to_center = data.pivot_table(
    index='locality_name', values=['cityCenters_nearest'], aggfunc=['min', 'max'])
location_to_center.columns = ['min_d', 'max_d']

location_to_airport = data.pivot_table(
    index='locality_name', values=['airports_nearest'], aggfunc=['min', 'max'])
location_to_airport.columns = ['min_d', 'max_d']

def find_locations(dist, pvt_tbl):
    if dist is None:
        return set(pvt_tbl.index)
    return set(pvt_tbl.query('min_d-100 <= @dist <= max_d+100').index)

def attempt_to_find_location(row):
    global location_to_center
    global location_to_airport
    not_stated = 'не указано'

    if row['locality_name'] is not np.NaN:
        return row

    center = row['cityCenters_nearest']
    airport = row['airports_nearest']
    if np.isnan(center) and np.isnan(airport):
        print(f'- Для NaN NaN невозможно найти локацию')
        #row['locality_name'] = not_stated # Оставим NaN для удаления
        return row
    res = (
        find_locations(center, location_to_center)
      & find_locations(airport, location_to_airport)
    )
    if len(res) == 0:
        print(f'- Для {center: 5.0f} {airport: 6.0f} не удалось найти локацию')
        row['locality_name'] = not_stated
        return row

    if len(res) > 1:
        print(f'- Для {center: 5.0f} {airport: 6.0f} неоднозначная локация: {res}')
        row['locality_name'] = not_stated
        return row
    
    res = res.pop()
    print(f'+ Для {center: 5.0f} {airport: 6.0f} определена локация: {res}')
    row['locality_name'] = res
    return row

data=data.apply(attempt_to_find_location, axis=1)
````

30 локаций удалось найти по расстояниям.   
Там, где определить населенный пункт однозначно не удалось,   
'locality_name' присвоено значение 'не указано' (11шт)    
Удалим строки без адреса и расстояний (8шт):

### Заполнение пропусков расстояний до центра и до аэропорта.  


Заменим пропущенные расстояния до центра и до аэропорта  
на средние величины для каждой локации:

````python
location_to_center = data.pivot_table(
    index='locality_name', values=['cityCenters_nearest'])
location_to_airport = data.pivot_table(
    index='locality_name', values=['airports_nearest'])

def fill_na_distances(row):
    global location_to_center
    global location_to_airport
    locality = row['locality_name']
    center   = row['cityCenters_nearest']
    airport  = row['airports_nearest']

    if np.isnan(center) and locality in location_to_center.index:
        print('c', end='')
        row['cityCenters_nearest'] = location_to_center.loc[locality, 'cityCenters_nearest']
    if np.isnan(airport) and locality in location_to_airport.index:
        print('a', end='')
        row['airports_nearest'] = location_to_airport.loc[locality, 'airports_nearest']
    return row

data = data.apply(fill_na_distances, axis=1)    
````

т.о. для подбора этажности дома достаточно линейки 5 9 12 16 и 25  
первый подходящий вариант можно использовать для заполнения:

````python
def fill_na_floors_total(row):
    if not np.isnan(row['floors_total']):
        return row
    for f in [5, 9, 12, 16, 25]:
        if f >= row['floor']:
            row['floors_total'] = f
            return row
    print("won't happen")
    return row

data = data.apply(fill_na_floors_total, axis=1)
data['floors_total'] = data['floors_total'].astype('uint8')
````

### Заполнение пропусков высоты потолков:

Расчитаем медиану высоты потолков:  

````python
ceiling_height_median = data['ceiling_height'].median()
ceiling_height_median
````

заполним пропуски медианой:

````python
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_height_median)
````

затем переведем в сантиметры и в тип int:

````python
data['ceiling_height'] = (data['ceiling_height']*100).astype('int16')
ceiling_height_median = int(ceiling_height_median * 100)
````

исправим артефакты:

````python
# потолки 100 и 200 см:
data['ceiling_height'] = data['ceiling_height'].where(data['ceiling_height']!=100, ceiling_height_median)
data['ceiling_height'] = data['ceiling_height'].where(data['ceiling_height']!=200, ceiling_height_median)
# потолки 100 метров:
data['ceiling_height'] = data['ceiling_height'].where(data['ceiling_height']!=10000, ceiling_height_median)
# потолки ниже 220 см
data['ceiling_height'] = data['ceiling_height'].where(data['ceiling_height']>=220, data['ceiling_height']+100)
# потолки выше 22 метров (вероятно ошиблись на порядок):
data['ceiling_height'] = data['ceiling_height'].where(data['ceiling_height']<2200, data['ceiling_height']//10)
# потолки 6 метров и выше - скорее всего ошибочные данные:
data['ceiling_height'] = data['ceiling_height'].where(data['ceiling_height']<600, ceiling_height_median)
````

````python
def create_conditions(col, df, show=True):
    '''Функция создает аргумент для DataFrame.query() выделяющий срез строк внутри "усов" boxplot
    
    Ключевые аргументы:
    -- col  имя столбца DataFrame с числовым значением
    -- df   ссылка на DataFrame
    -- show выводить ли на экран интервал значений (по умолчанию True)
    '''
    q1 = df[col].quantile(q=0.25)
    q3 = df[col].quantile(q=0.75)
    w = (q3-q1)*1.5
    if show: print(col+f': [{q1-w:.2f} ... {q3+w:.2f}]')
    return '('+str(q1-w)+'<='+col+'<='+str(q3+w)+')'
````

и расчитаем среднюю стоимость квадратного метра по километрам:

````python
km_sqm = km_pivot[('sum', 'last_price')] / km_pivot[('sum', 'total_area')]
km_sqm_df = pd.DataFrame(km_sqm, columns=['mean_price_sqm'])
#km_sqm_df         ✔
````
